```python
import pandas as pd
import numpy as np
import pylab as pl
%pylab inline
```

```python
summer = pd.read_csv('Summer_2018_Load.csv')
```

```python
summer.columns
```

```python
summer.head()
```

```python
summer['Timestamp'] = pd.to_datetime(summer['Timestamp'], format='%H:%M:%S').dt.time
```

```python
plt.figure()
fig, ax1 = plt.subplots(figsize=(10,4))

plt.xticks(rotation=50)
ax1.plot(summer['Timestamp'], summer['Average Power(kW)'], color ='orange')
ax1.set_ylabel('Average Power (kW)')

ax2 = ax1.twinx()
ax2.plot(summer['Timestamp'], summer['Customer Load'], color ='g')
ax2.set_ylabel('Customer Load(kW)')
```
